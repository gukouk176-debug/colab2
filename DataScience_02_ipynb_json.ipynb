## 制御文： for文 繰り返し処理

```python
#for文 繰り返し　//for ループ変数 in 繰り返し条件オブジェクト:

print("0～10（の手前）まで繰り返して表示")
for i in range(10):
    print(i)

print() #改行

# 改行せずに，　文字の間に「，」を挿入
print('3～16（の手前）まで繰り返して表示')
for i in range(3,16):
    print(i, end = ', ')

print()

print("1～11（の手前）まで2つ飛ばしで表示")
for i in range(1, 11, 2):
  print(i, end = '  ')
```

## 制御文： while の使い方

```python
num = 0

print('Start')

while num < 2:
    print("num = " + str(num))
    num += 1

print('End')

print('The final value is ', num)
```

### 演習4： while 文の練習
+ breakで何が起こっているか考えてください

```python
numbers = [2, 15, 44, 34, 52, 23, 62, 49, 70]

i = 0

while i < len(numbers):
    if numbers[i] > 50:
        print('More than 50 has been found!')
        break
    i += 1
```

### 演習5： while 文の練習
+ continue で何が起こっているか考えてください

```python
numbers = [2, 15, 44, 34, 52, 23, 62, 49, 70]

i = 0

while i < len(numbers):
    if numbers[i] > 50:
        i += 1
        continue
    print(numbers[i])
    i += 1
```

## 演習１：変数の練習

```python
#変数の練習
#(1)変数fに15を代入、変数gに30を代入して変数hにfとgの積の値を代入して表示する
f = 15
g = 30
h = f*g
print('h =',h)
#(2)変数iに300を代入、変数jに変数hから変数iの割り算の余りの値を代入して表示する
i = 300
j = h % i
print('j =',j)
```

## 演習２：if文の練習

```python
#if文の練習
#(4)aの2倍の値とbを比べてaの2倍が大きければ「aの2倍が大きい」と出力、小さければ「aの2倍は小さい」と出力
a = 3
b = 3
if a*2 > b:
    print('aの2倍が大きい')
else:
    print('aの2倍がbより小さい')

#(5)aとbの積が10以上ならば「abは大きい」と出力，5以上10未満なら「abは普通」,5未満ならば「abは小さい」と出力

if a * b >= 10:
    print('abは大きい')
elif 5 <= a*b < 10:
    print('abは普通')
else:
    print('abは小さい')
```

```python
#for文の練習
#(6)for文を使って２，４，６，８，１０を出力する
for i in range(1,6):
    print(2*i)

print()
#(7)for文を使って１５，１６，１７，１８，１９，２０，２１を表示
for i in range(15,22):
    print(i)

print()
#(8)for文とif文を使って以下のリストの数字で3の倍数のものだけ表示する
list =[3,5,8,22,35,36,21,90,89,88]
for i in list:
    if i%3==0:
        print(i)
```
