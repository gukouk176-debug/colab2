## NumPy入門
+ 表形式のデータ間の計算を簡単にできるツール群

```python
#配列,ライブラリnumpy使用
#使用するライブラリの宣言import ※asはライブラリを使いやすいように名前の置き換え
import numpy as np

#配列の宣言．　ndarrayを作成
array1 = np.array([1,1,1])
array2 = np.array([3,3,5])

print('array1 =', array1)
print('array2 =', array2)
```

+ 配列の次元

```python
# 配列の形状（各次元のサイズ）
print(array1.shape)
```

```python
# 配列の「最初の」次元のサイズ
print(len(array1))
```

+ 配列の演算

```python
#配列の演算と表示
array3 = array1 + array2
array4 = array1*array2    # 要素ごとの乗算（内積ではない）
array5 = array4 - array3

print('array4 =', array4)
print('array5 =', array5)
```

```python
# 内積
np.dot(array1, array2)
```

```python
# ２次元行列
matrix1 = np.array([[2,1,3],[5,4,8],[3,3,2]])
matrix2 = np.array([[5,5,1],[6,1,2],[9,3,7]])

print('matrix1 =\n', matrix1)
print('matrix2 =\n', matrix2)
```

```python
# 配列の形状
print(matrix1.shape)
```

```python
# 2次元行列の計算と表示
matrix3 = matrix1+matrix2
matrix4 = matrix1*matrix2
print('matrix3 =\n', matrix3)
print('matrix4 =\n', matrix4)

#配列の各行,　列の抜き出し
print('matrix4の1行目', matrix4[0])
print('matrix4の2行目', matrix4[1])
print('matrix4の3行目', matrix4[2])

print('matrix4の1行1列目', matrix4[0][0])
print('matrix4の2行3列目', matrix4[1][2])
```

```python
# ２次元行列の内積
print('inner product =\n', np.dot(matrix1, matrix2))
```

```python
# 配列の「最初の」次元のサイズ
matrix5 = np.array([[2,1,3],[5,4,8],[3,3,2],[4,3,2]])
print(len(matrix5))
```

```python
# データ型
array1.dtype
```

```python
# 次元数と要素数
print('array1の次元数：', array1.ndim)
print('array1の要素数：', array1.size)

print('matrix1の次元数：', matrix1.ndim)
print('matrix1の要素数：', matrix1.size)
```

```python
#配列の1次元化
print('matrix1 = \n', matrix1)
one_dim_array = matrix1.reshape(-1,)
print('One dimensional array: ', one_dim_array)

#1次元配列の多次元化,変換できない多次元形状だとエラーが出るので注意
#例）9個の1次元配列を２次元化配列にしようとした場合、２次元化できない要素が出てくるためエラーとなる
two_dim_array = matrix1.reshape(3,3)
print('3 x 3 martirx = \n', two_dim_array)

#単なるリスト作成の場合
List =[1,3,6,7]
print('List =',List)
print(List[0])
```

## Matplotlib入門

```python
# Matplotlib のインポート
import matplotlib as mpl
import matplotlib.pyplot as plt

# Seaborn のインポート（より詳しくは次のセクション参照）
import seaborn as sns

np.random.seed(0)         # 乱数シードの固定
x = np.random.randn(30)  # 平均0，標準偏差1の正規分布からの30個の乱数
y = np.random.rand(30)   # 0以上，1未満の小数の30個の乱数

plt.figure(figsize=(15,6))

#plt.plot(x, y, 's')
plt.scatter(x, y, s = 50, c = 'red', alpha = 0.5, linewidths = 1, edgecolors = 'blue', marker = '^')
```

```python
plt.figure(figsize=[15,6])

# 2行1列のグラフの1つ目
plt.subplot(2,1,1)
x = np.linspace(-10,10, 200)
plt.plot(x, np.sin(x))

# 2つ目
plt.subplot(2,1,2)
y = np.linspace(-10, 10, 100)
plt.plot(y, 2*np.sin(2*y), c = 'purple')

plt.grid(True)
```

```python
np.random.seed(0)

plt.figure(figsize=(15,6))

plt.subplot(2,1,1)
# randで0～1の頻度，ビンの数を50個
plt.hist(np.random.rand(10**5), bins = 50, range = (0,1))

plt.grid(True)

plt.subplot(2,1,2)
# randnで正規分布（平均を50に変更，値を10倍），ビンの数を50個
plt.hist(np.random.randn(10**5)*10 + 50, bins = 50, range = (0,100))

plt.grid(True)
```

## Seaborn 入門
「An introduction to seaborn」にあるチュートリアルからいくつかを演習問題としています


```python
# チップの金額のデータセットについての散布図の例

# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# 回帰モデルの可視化とグリッドによる表示

sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
# ヒストグラムを作成

sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
# カテゴリーデータ（smoker/non-smoker）の表示

sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```
